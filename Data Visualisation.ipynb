```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df= pd.read_csv('googleplaystore.csv')
df= df.dropna() #removes all null values
df['Reviews'] = pd.to_numeric(df['Reviews'] ,errors= 'coerce') #conerts to numeric data
df['Installs']= df['Installs'].str.replace('+' , '').str.replace(',', '').astype(int)
df['Price'] = df['Price'].str.replace('$' , '').astype(float)
df['LastUpdated']= pd.to_datetime(df['LastUpdated'])
df= df.drop_duplicates()
def convert_size(size):
    if 'M' in size:
        return float(size.replace('M' ,''))
    elif 'k' in size:
        return float(size.replace('k', ''))/1000 #kilo bytes converted to mega bytes
    else:
        return np.nan

df['Size'] = df['Size'].apply(lambda x: convert_size(x) if isinstance(x, str) else x)
category_data= df.groupby('Category')[['Rating', 'Reviews' , 'Installs']].mean().round(2)
correlation= df[['Rating', 'Reviews', 'Size', 'Installs', 'Price']].corr()
plt.figure(figsize= (10,5))
sns.heatmap(correlation, annot=True, cmap= 'BuPu') #crest, coolwarm, cubehelix, greens, oranges,reds etc)
plt.title('CORRELATION HEATMAP')
plt.show()
```

```python
#BAR PLOT 
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x='Category' , y='Installs', data= df, estimator=np.mean, errorbar= ('ci', 100))
plt.xticks(rotation=90)
plt.title('Average Installs by Category')
plt.show()
```

```python
avg_installs_by_type= df.groupby(['Category', 'Type'])['Installs'].mean().reset_index()
plt.figure(figsize=[10,10])
sns.barplot(x= 'Category' , y='Installs', hue= 'Type', data=avg_installs_by_type, palette='pastel')
plt.title("avg number of installs by category")
plt.xticks(rotation=90)
plt.xlabel('avg number of installs')
plt.ylabel('category')
plt.legend(title= 'Type')
plt.show()
```

```python
#LINE PLOT
```

```python
#convert last updated to datetime
df['LastUpdated']= pd.to_datetime(df['LastUpdated'])
#extract year from lastUpdated
df['Year']= df['LastUpdated'].dt.year
#calculate the average rating per year
average_rating_per_year= df.groupby('Year')['Rating'].mean().reset_index()

#plot a lineplot
plt.figure(figsize=[10,10])
sns.lineplot(x='Year', y='Rating', data= average_rating_per_year,marker='o')
#add a trend line
sns.regplot(x='Year', y='Rating', data= average_rating_per_year, scatter= False, color='r')
#add labels and title
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.title('average rating of apps over time')
plt.legend()
plt.grid(True)
plt.show()
```
